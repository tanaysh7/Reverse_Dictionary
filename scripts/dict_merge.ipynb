```python
import io
k=io.open('wordlist.txt')
import json
```

```python
data = '{'
c=0
for row in k:
    wrd=row.rstrip('\n')
    data+= '"'+wrd+'" :{'
    synonyms = []
    antonyms = []
    wordlist=[]
    definitions=[]
    synonyms = []
    antonyms = []
    partofspeech=[]
    for ss in wordnet.synsets(wrd):
        wordlist.append(ss.name().split('.')[0])
        definitions.append(ss.definition())
        partofspeech.append(ss.pos())
        for l in  ss.lemmas():
            synonyms.append(l.name())
            if l.antonyms():
                antonyms.append(l.antonyms()[0].name())
    synonyms=list(set(synonyms))
    antonyms=list(set(antonyms))
    data+='"words" :'
    data+=json.dumps(wordlist)
    data+=',"definitions" :'
    data+=json.dumps(definitions)
    data+=',"partofspeech" :'
    data+=json.dumps(partofspeech)
    data+=',"synonyms" :'
    data+=json.dumps(synonyms)
    data+=',"antonyms" :'
    data+=json.dumps(antonyms)
    data+='},'
    #if c > 10: break
    c+=1
data=data.rstrip(',');
data+="}"
        
```

```python
jsondict = open("jsondict.json", 'w')
jsondict.write(data)
jsondict.close()
```

```python

from nltk.tokenize import regexp_tokenize
from nltk.corpus import wordnet as wn
from nltk.corpus import stopwords
import json
```

```python
 nltk.download('stopwords')
```

```python
ensw = stopwords.words('english')
load the word_dict

with open("jsondict.json", "rb") as file:
    dictionary = file.read()

dictionary = json.loads(dictionary)

for key in dictionary:
    print(dictionary.get(key))
    break
#make the dictionary consisting of words and all the definition indices it is present in

definition_dictionary = {}
```

```python
for dict_word, info in dictionary.items():
    for definition in info.get('definitions'):
        word_array = regexp_tokenize(definition, "[\w']+")
        word_array = [item for item in word_array if item not in ensw]
        for word in word_array:
            if definition_dictionary.get(word):
                definition_dictionary[word].append(dict_word)
            else:
                definition_dictionary.update({word:[dict_word]})

for key in definition_dictionary:
    print(key)
    print(definition_dictionary.get(key))
    break

definition_dictionary_data = json.dumps(definition_dictionary).encode("utf-8")

with open("definition_dictionary.json", "wb") as file:
    file.write(definition_dictionary_data)
```

```python
searchstr="plant life cycle"
searchlist=[]
for strn in searchstr.split(" ") : 
    searchlist+=definition_dictionary.get(strn)
#dictionar of words and their relevance(total counts)
searchdict=dict((x,searchlist.count(x)) for x in set(searchlist))
```

```python
sorteddict=sorted(searchdict.iteritems(), key=lambda (k,v): (v,k),reverse =True)
```
